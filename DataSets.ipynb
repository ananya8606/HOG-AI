# DataSet Generation

##### It is a process of creating training and validation data for a neural network.We define the inputs and the corresponding correct output of all possible cases that we can generate.

## DataSet

#####  Numpy arraya in general consisting of input valus in one and corresponding output values shaped according to the network architecture in another.

### Xor Problem

##### In this we solve xor operation using a neural network .It is the simplest example of use of ML. We have two inputs and one output.

```python
# We will create a simple dataset for xor problem.

import numpy as np

#packing input data
a=[0,0,1,1]
b=[0,1,0,1]

d=[]
for i in range(0,4):
    t=[a[i],b[i]]
    s=np.array(t)
    d.append(s)
    
data=np.array(d)

c=[0,1,1,0]
out=np.array(c)

print(data)
print(data.shape)

print(out)
print(out.shape)

location=input("Enter the path of location to save the data:")

np.savez_compressed(location+"\datafile",x=data,y=out)
```

```python
# We will now retrieve the created file and print its value

import numpy as np

location=input("Enter the path of location to get the data:")
print(location)

data=np.load(location+"\datafile.npz")
print(data['x'])
print(data['y'])
```

### In the below code we will plot loss and accuracy of training of our CNN( convolutional neural network )

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Activation,Dense
import matplotlib
import matplotlib.pyplot as plt

location=input("Enter the path of location to get the data:")
print(location)

tmp=np.load(location+"\datafile.npz")

data=tmp['x']
out=tmp['y']

model = Sequential()
model.add(Dense(16, input_shape=(2,)))
model.add(Activation('relu'))
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dense(16))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()

history=model.fit(data,out,epochs=500)

plt.plot(history.history['accuracy'])
plt.title("Model Accuracy")
plt.show()

plt.plot(history.history['loss'])
plt.title("Model loss")
plt.show()
```
